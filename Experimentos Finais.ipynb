# Redução de dimensionalidade

```python
from sklearn import preprocessing
```

```python
s_values = preprocessing.scale(df)
print(s_values.shape)
s_values[0]
```

## PCA

```python
from sklearn.decomposition import PCA
```

#### Vamos calcular vários PCAs para números diferentes de componentes (de 100 até 400)

```python
import time
start_time = time.time()
# %%time
pcas = []
pcas_data = []

for i in range(1, 5):
    pcas.append(PCA(n_components=i*100, random_state=42))
    pcas[-1].fit(s_values)
    pcas_data.append(pcas[-1].transform(s_values))
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

## Latent Semant Indexing (LSI)

Vamos aplicar o LSI, usando o Truncated SVD nos dados de termo-frequencia do TF-IDF

```python
from sklearn.decomposition import TruncatedSVD
```

```python
import time
start_time = time.time()

lsis = []
lsis_data = []

for i in range(1, 5):
    lsis.append(TruncatedSVD(n_components=i*100, random_state=42))
    lsis[-1].fit(df)
    lsis_data.append(lsis[-1].transform(df))
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

## Non-Negative Matrix Factorization

```python
from sklearn.decomposition import NMF
```

```python
import time
start_time = time.time()

nmfs = []
nmfs_data = []

for i in range(1, 5):
    nmfs.append(NMF(n_components=i * 100, init='random', random_state=42, max_iter=2000))
    nmfs[-1].fit(df)
    nmfs_data.append(nmfs[-1].transform(df))
    print(i)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

# Agrupamento

## K-means

```python
import numpy as np
from sklearn.cluster import KMeans
```

O K-Means usa a distância Euclidiana e no sklearn não pode ser alterada. Para usar outra medida de distância podemos usar o K-Medoids.

```python
from sklearn.cluster import KMeans
```

```python
min_clusters = 2
max_clusters = 40
```

```python
%%time

km = []

i = min_clusters

while i < max_clusters:
    km.append(KMeans(n_clusters=i, random_state=42))
    km[-1].fit_predict(df)
    i += 1
```

#### K-Means em PCAs de 100 a 400 componentes

```python
import time
start_time = time.time()

pcas_km = []

for unit_pca_data in pcas_data:
    aux = []
    for j in range(min_clusters, max_clusters):
        aux.append(KMeans(n_clusters=j, random_state=42))
        aux[-1].fit_predict(unit_pca_data)
    pcas_km.append(aux)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

#### K-Means em LSIs de 100 a 400 componentes

```python
import time
start_time = time.time()

lsis_km = []

for unit_lsi_data in lsis_data:
    aux = []
    for j in range(min_clusters, max_clusters):
        aux.append(KMeans(n_clusters=j, random_state=42))
        aux[-1].fit_predict(unit_lsi_data)
    lsis_km.append(aux)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

#### K-Means em NMF com 100 a 400 componentes

```python
import time
start_time = time.time()

nmfs_km = []

for unit_nmf_data in nmfs_data:
    aux = []
    for j in range(min_clusters, max_clusters):
        aux.append(KMeans(n_clusters=j, random_state=42))
        aux[-1].fit_predict(unit_nmf_data)
    nmfs_km.append(aux)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

#### DBSCAN no df original

```python
eps = [0.25, 1.5]
min_samples = [10, 11]

dbscan = []
pred = []

i = eps[0]
while(i <= eps[1]):
    for j in range (min_samples[0], min_samples[1]):
        dbscan.append(DBSCAN(eps = i, min_samples = j))
        pred.append(dbscan[len(dbscan) - 1].fit_predict(df))
    i += 0.25
    
for i in range(0, len(dbscan)):
    print(dbscan[i], pred[i])
```

Não funciona! Apenas encontramos -1s e 0s, assim não agrupando nada.

#### DBSCAN no PCA de 100 a 400 componentes

```python
%%time

eps = [0.25, 5]
min_samples = [3, 11]
eps_increase = 0.25

pcas_db = []

for unit_pca_data in pcas_data:
    aux=[]
    i = eps[0]
    while(i <= eps[1]):
        for j in range (min_samples[0], min_samples[1]):
            aux.append(DBSCAN(eps = i, min_samples = j))
            aux[-1].fit_predict(unit_pca_data)
        i += eps_increase
    pcas_db.append(aux)
    
for unit_pca_db in pcas_db:
    for db in unit_pca_db:
        counter = Counter(db.labels_)
        if len(counter) > 3:
            print(db, counter, len(counter))
```

Nenhum dado interessante! Não encontrei nenhum que consegue criar mais de 2 agrupamentos, e a maioria são outliers (-1)

#### DBSCAN em LSI de 100 a 400 componentes

```python
import time
start_time = time.time()

eps = [0.15, 0.53]
min_samples = [3, 11]
eps_increase = 0.01

lsis_db = []

for unit_lsi_data in lsis_data:
    aux=[]
    i = eps[0]
    while(i <= eps[1]):
        for j in range (min_samples[0], min_samples[1]):
            aux.append(DBSCAN(eps = i, min_samples = j))
            aux[-1].fit_predict(unit_lsi_data)
        i += eps_increase
    lsis_db.append(aux)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()

acceptable_lsis_db = []
    
for unit_lsi_db in lsis_db:
    aux = []
    for db in unit_lsi_db:
        counter = Counter(db.labels_)
        if counter[-1] <= 800 and len(counter) > 3:
#             print(db, counter, len(counter))
            aux.append(db)
    acceptable_lsis_db.append(aux)
    
acceptable_lsis_db
```

```python
Counter(acceptable_lsis_db[0][0].labels_)
```

#### DBSCAN em NMFs de 100 a 400 componentes

```python
import time
start_time = time.time()

eps = [0.01, 0.20]
min_samples = [3, 11]
eps_increase = 0.01

nmfs_db = []

for data in nmfs_data:
    aux=[]
    i = eps[0]
    while(i <= eps[1]):
        for j in range (min_samples[0], min_samples[1]):
            aux.append(DBSCAN(eps = i, min_samples = j))
            aux[-1].fit_predict(data)
        i += eps_increase
    nmfs_db.append(aux)
    print("--- %s seconds ---" % round((time.time() - start_time), 3))
    start_time = time.time()

acceptable_nmfs_db = []
    
for unit_db in nmfs_db:
    aux = []
    for db in unit_db:
        counter = Counter(db.labels_)
        if counter[-1] <= 800 and len(counter) > 3:
            print(db, counter, len(counter))
            aux.append(db)
    acceptable_nmfs_db.append(aux)
```
